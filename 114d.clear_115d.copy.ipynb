d.clear

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
greats.clear()
print(greats)
```

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
del greats
print(greats)
```

d.copy

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
buffer = greats.copy()
del buffer['관우']
print(f'greats:{greats}')
print(f'buffer:{buffer}')
```
