d.pop

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
level = greats.pop('관우')
print(f'greats:{greats}|nlevel:{level}')
```

```python
level = greats.pop('제갈량',99)
print(f'greats:{greats}|제길량의 레벨:{level}')
```

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
while len(greats) > 0:
  general = greats.popitem()
  print(f'추출한 장군:{general}|greats 상태:{greats}')
  print('-'*50)
```

d.update

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
others = {'제갈량':99,'황충':70}
greats.update(others)
print(greats)
```

```python
greats.update(유비=90,관우=96)
print(greats)
```
