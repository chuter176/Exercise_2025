d1 | d2

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
others = {'제갈량':99,'황충':70}
updated = greats|others
print(updated)
```

```python
updated2 = updated|{'유비':90,'관우':96}
print(updated2)
```

```python
greats |= others
print(greats)
```

sorted

```python
greats = {'유비':89,'관우':95,'장비':91,'조운':94}
ret = sorted(greats)
print(ret)
```

```python
ret = sorted(greats.items())
print(ret)
```

```python
ret = sorted(greats.items(),reverse=True)
print(ret)
```

```python
ret = sorted(greats.items(),key=lambda x:x[1])
print(ret)
```

```python
ret = sorted(greats.items(),key=lambda x:x[1],reverse=True)
print(ret)
```

```python
for name, level in ret:
    print(f'<{name}>의 레벨은 <{level}>입니다.')
```
