```python
def makeForm(func):
  def drawLine():
    print('-'*50)
    func()
    print('-'*50)
  return drawLine
```

```python
report_new = makeForm(report)
report_new()
```

```python
def decoFunc(func):
  def wrapper(*args, **kwargs):
    func(*args, **kwargs)
  return wrapper
```
